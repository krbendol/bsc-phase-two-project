```python
df.columns = df.columns.str.replace(' ', '')
df.columns= df.columns.str.lower()
```

```python
df['winter_spring_summer_fall'] = df['date'].str[0:2]
```

```python
df['winter_spring_summer_fall'] = df['winter_spring_summer_fall'].replace(['1/'], '1')
df['winter_spring_summer_fall'] = df['winter_spring_summer_fall'].replace(['2/'], '1')
df['winter_spring_summer_fall'] = df['winter_spring_summer_fall'].replace(['3/'], '2')
df['winter_spring_summer_fall'] = df['winter_spring_summer_fall'].replace(['4/'], '2')
df['winter_spring_summer_fall'] = df['winter_spring_summer_fall'].replace(['5/'], '2')
df['winter_spring_summer_fall'] = df['winter_spring_summer_fall'].replace(['6/'], '3')
df['winter_spring_summer_fall'] = df['winter_spring_summer_fall'].replace(['7/'], '3')
df['winter_spring_summer_fall'] = df['winter_spring_summer_fall'].replace(['8/'], '3')
df['winter_spring_summer_fall'] = df['winter_spring_summer_fall'].replace(['9/'], '4')
df['winter_spring_summer_fall'] = df['winter_spring_summer_fall'].replace(['10'], '4')
df['winter_spring_summer_fall'] = df['winter_spring_summer_fall'].replace(['11'], '4')
df['winter_spring_summer_fall'] = df['winter_spring_summer_fall'].replace(['12'], '1')
```

```python
df['winter_spring_summer_fall'] = df['winter_spring_summer_fall'].astype(int)
```

```python
df['sqft_basement'] = pd.to_numeric(df.sqft_basement, errors='coerce')
```

```python
df['basement'] = df['sqft_basement'] > 0
```

```python
df['log_price'] = np.log(df['price'])
#df['log_price'] = np.power((df['price']),1/10)
```

```python
df = df.drop(columns=['date', 'sqft_basement'])
```

```python
df = df.dropna()
```

```python
df
```

```python
X = df[['bathrooms', 'sqft_living', 'floors', 'basement', 'grade', 'bedrooms', 'condition', 'view', 'waterfront', 'sqft_living15', 'sqft_lot15', 'sqft_lot', 'yr_renovated', 'yr_built', 'winter_spring_summer_fall', 'lat', 'long']]
y = df['log_price']
```

```python
slr = LinearRegression()
slr.fit(X, y)
```

```python
print(slr.intercept_)
print(slr.coef_)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=5)
print(len(X_train), len(X_test), len(y_train), len(y_test))
```

```python
mlr_split = LinearRegression()

#fit
mlr_split.fit(X_train, y_train)

# calculate predictions on training and test sets
y_hat_train = mlr_split.predict(X_train)
y_hat_test = mlr_split.predict(X_test)

#compare MSE for the predicted training and test values 
train_mse = mean_squared_error(np.exp(y_train), np.log(y_hat_train))
test_mse = mean_squared_error(np.exp(y_test), y_hat_test)
print('Train Mean Squarred Error:', train_mse)
print('Test Mean Squarred Error:', test_mse)

print('Train score:', mlr_split.score(X_train, y_train))
print('Test score:', mlr_split.score(X_test, y_test))
```

```python
def get_score(prediction, lables):    
    print('R2: {}'.format(r2_score(prediction, lables)))
    print('RMSE: {}'.format(np.sqrt(mean_squared_error(prediction, lables))))
```

```python
def train_test(estimator, x_trn, x_tst, y_trn, y_tst):
    prediction_train = estimator.predict(x_trn)
    # Printing estimator
    print(estimator)
    # Printing train scores
    get_score(prediction_train, y_trn)
    prediction_test = estimator.predict(x_tst)
    # Printing test scores
    print("Test")
    get_score(prediction_test, y_tst)
```

```python
GBest = ensemble.GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05, max_depth=3, max_features='sqrt',
                                               min_samples_leaf=15, min_samples_split=10, loss='huber').fit(X_train, y_train)
train_test(GBest, X_train, X_test, y_train, y_test)
```

```python
scores = cross_val_score(GBest, X_test, y_test, cv=5)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
